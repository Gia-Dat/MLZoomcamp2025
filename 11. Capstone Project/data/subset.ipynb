```python
import os
import shutil
import random

## Translate name
translate = {
    "cane": "dog", "cavallo": "horse", "elefante": "elephant", 
    "farfalla": "butterfly", "gallina": "chicken", "gatto": "cat", 
    "mucca": "cow", "pecora": "sheep", "scoiattolo": "squirrel", 
    "ragno": "spider"
}

# Path
original_dataset_path = '/kaggle/input/animals10/raw-img'
subset_path = 'animal_subset'

if not os.path.exists(subset_path):
    os.makedirs(subset_path)

for italian_name in os.listdir(original_dataset_path):
    class_dir = os.path.join(original_dataset_path, italian_name)
    
    if os.path.isdir(class_dir):
        english_name = translate.get(italian_name, italian_name)
        images = [f for f in os.listdir(class_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        # Choose random 10 images
        n_sample = min(len(images), 10)
        sample_images = random.sample(images, n_sample)
        
        # Create a new folder
        target_class_dir = os.path.join(subset_path, english_name)
        os.makedirs(target_class_dir, exist_ok=True)
        
        # Copy images
        for img in sample_images:
            shutil.copy(os.path.join(class_dir, img), os.path.join(target_class_dir, img))

print(f"'{subset_path}' created")
```

```python
shutil.make_archive('animal_subset', 'zip', 'animal_subset')
```

## Set up framework

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size = 0.2, random_state = 1)
df_train, df_val = train_test_split(df_full_train, test_size = 0.25, random_state = 1)
```

```python
len(df_full_train), len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)

y_train = df_train.target.values
y_val = df_val.target.values
y_test = df_test.target.values

del df_train['target']
del df_val['target']
del df_test['target']
```

## Train model

```python
categorical
```

```python
numerical = [col for col in df_train.columns if col not in categorical]
numerical
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import KFold
from sklearn.pipeline import make_pipeline

import numpy as np
import warnings # ignore warnings of the Logistic Regression later
warnings.filterwarnings("ignore")
```

```python
train_dicts = df_train.to_dict(orient = 'records')
dv = DictVectorizer(sparse = False)
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val.to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
# Logistic Regression
model_lr = LogisticRegression(max_iter = 2000, random_state = 1)
model_lr.fit(X_train, y_train)
y_pred_lr = model_lr.predict_proba(X_val)[:, 1]
print("Logistic Regression Model:", roc_auc_score(y_val, y_pred_lr))

# Decision Tree
model_dt = DecisionTreeClassifier(random_state = 1)
model_dt.fit(X_train, y_train)
y_pred_dt = model_dt.predict_proba(X_val)[:, 1]
print("Decision Tree Model:", roc_auc_score(y_val, y_pred_dt))

# Random Forest
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict_proba(X_val)[:, 1]
print("Random Forest Model:", roc_auc_score(y_val, y_pred_rf))

# XGBoost
model_xgb = xgb.XGBClassifier(random_state=42)
model_xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict_proba(X_val)[:, 1]
print("XGBoost Model:", roc_auc_score(y_val, y_pred_xgb))
```

- It looks like Random Forest do the best job out of 4 models.
- Now I will tuning all these models and see how well they could be

```python
# Tuning C for the Logistic Regression
print("==== Logistic Regression Tuning ====")
for C in [0.01, 0.1, 1, 3, 5, 10]:
    model_lr = LogisticRegression(C=C, max_iter=2000, random_state=1)
    model_lr.fit(X_train, y_train)
    
    y_pred = model_lr.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    print(f"C={C:<4}  AUC={auc:.4f}")
```

With C=3, LR Model do the best

```python
print("\n==== Decision Tree Tuning ====")
for depth in [3, 5, 7, 10, None]:
    for leaf in [1, 3, 5, 10]:
        model_dt = DecisionTreeClassifier(
            max_depth=depth,
            min_samples_leaf=leaf,
            random_state=1
        )
        model_dt.fit(X_train, y_train)
        
        y_pred = model_dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        print(f"depth={str(depth):<4}  leaf={leaf:<2}  AUC={auc:.4f}")
```

With max_depth = 5 & min_samples_leaf = 10, DT Model do the best

```python
print("\n==== Random Forest Tuning ====")
for n in [100, 200, 300]:
    for depth in [None, 5, 10, 20]:
        for leaf in [1, 3, 5]:
            model_rf = RandomForestClassifier(
                n_estimators=n,
                max_depth=depth,
                min_samples_leaf=leaf,
                random_state=42,
                n_jobs=-1
            )
            model_rf.fit(X_train, y_train)
            
            y_pred = model_rf.predict_proba(X_val)[:, 1]
            auc = roc_auc_score(y_val, y_pred)
            print(f"trees={n:<3} depth={str(depth):<4} leaf={leaf:<2} → AUC={auc:.4f}")
```

With n_estimators = 200, max_depth = None, mean_sample_leaf = 1, RF do the best

```python
print("\n==== XGBoost Tuning ====")
for eta in [0.01, 0.05, 0.1, 0.3]:
    for depth in [3, 5, 7]:
        for child in [1, 3, 5, 7]:
            model_xgb = xgb.XGBClassifier(
                eta=eta,
                max_depth=depth,
                min_child_weight=child,
                n_estimators=300,
                subsample=0.8,
                colsample_bytree=0.8,
                random_state=42,
                eval_metric='logloss'
            )
            model_xgb.fit(X_train, y_train)

            y_pred = model_xgb.predict_proba(X_val)[:, 1]
            auc = roc_auc_score(y_val, y_pred)
            print(f"eta={eta:<4} depth={depth:<2} child={child:<2} → AUC={auc:.4f}")
```

With eta=0.1, max_depth=5, min_child_weight=3, XGB do the best

## Train the model with parameters tuning

```python
# Logistic Regression
model_lr = LogisticRegression(max_iter = 2000, random_state = 1, C=3)
model_lr.fit(X_train, y_train)
y_pred_lr = model_lr.predict_proba(X_val)[:, 1]
print("Logistic Regression Model:", roc_auc_score(y_val, y_pred_lr))

# Decision Tree
model_dt = DecisionTreeClassifier(
            max_depth=5,
            min_samples_leaf=10,
            random_state=1
        )
model_dt.fit(X_train, y_train)
y_pred_dt = model_dt.predict_proba(X_val)[:, 1]
print("Decision Tree Model:", roc_auc_score(y_val, y_pred_dt))

# Random Forest
model_rf = RandomForestClassifier(
                n_estimators=200,
                max_depth=None,
                min_samples_leaf=1,
                random_state=42,
                n_jobs=-1
            )
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict_proba(X_val)[:, 1]
print("Random Forest Model:", roc_auc_score(y_val, y_pred_rf))

# XGBoost
model_xgb = xgb.XGBClassifier(
                eta=0.1,
                max_depth=5,
                min_child_weight=3,
                n_estimators=300,
                subsample=0.8,
                colsample_bytree=0.8,
                random_state=42,
                eval_metric='logloss'
            )
model_xgb.fit(X_train, y_train)
y_pred_xgb = model_xgb.predict_proba(X_val)[:, 1]
print("XGBoost Model:", roc_auc_score(y_val, y_pred_xgb))
```

It look much better than previous and RF Model still the best model to use for this dataset
